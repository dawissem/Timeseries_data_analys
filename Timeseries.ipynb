```python
# Convertir la colonne 'Consumption' en numérique
data['Consumption'] = pd.to_numeric(data['Consumption'], errors='coerce')

# S'assurer que 'day_name' est bien dans les colonnes
data['day_name'] = pd.to_datetime(data['Date']).dt.day_name()

# Grouper les données par 'day_name' et calculer la somme de 'Consumption' uniquement
groupday = data.groupby('day_name')['Consumption'].sum().sort_values()

# Définir l'ordre des jours de la semaine
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Créer le graphique à barres
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=groupday.index, y=groupday.values, order=order)

# Afficher le graphique
plt.show()
```
